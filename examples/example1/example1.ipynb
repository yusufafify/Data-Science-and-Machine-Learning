### Example 1: Data Processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('./diabetes.csv')
dataset.replace({'Yes':1,'No':0},inplace=True)
dataset
```

```python
dataset.describe()
```

```python
dataset.info()
```

### For Practice: Introduce Null in the dataset and fix it

```python
dataset.iloc[10,4]=np.nan
dataset.iloc[5,4]=np.nan
dataset.iloc[8,4]=np.nan
dataset.iloc[11,4]=np.nan
dataset.iloc[9,0]=np.nan
```

```python
dataset.info()
```

```python
dataset['Gender'].replace({'Male':1,'Female':0},inplace=True)
dataset
```

```python
dataset.describe()
```

```python
dataset.hist(figsize=(10,10))
```

```python
dataset['Age'].fillna(dataset['Age'].mean(),inplace=True)
dataset['Age']=dataset['Age'].astype(int)
dataset.info()
```

```python
dataset.dropna(subset=['sudden weight loss'],inplace=True)
dataset.info()
```

### Normalization (eg: Normalize age)

```python
dataset['Age']=dataset['Age']/dataset['Age'].mean()
dataset
dataset.hist(figsize=(10,10))
```
