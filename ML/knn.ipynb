```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#sklearn library
from sklearn import datasets,model_selection, neighbors, metrics, preprocessing
```

### KNN (classification)

![KNN Illustration](../images/knn_illustration.png)

The K-Nearest Neighbors (KNN) algorithm is a simple, instance-based learning algorithm that classifies data points based on the class of their k nearest neighbors in the feature space.

```python
x, y= datasets.load_iris(return_X_y=True)
```

```python
x_normalize = preprocessing.StandardScaler()
x_norm = x_normalize.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x_norm,y,test_size=0.2,random_state=42,stratify=y)
```

```python
knn_model = neighbors.KNeighborsClassifier(n_neighbors=10)

knn_model.fit(x_train,y_train)
```

```python
y_pred = knn_model.predict(x_test)
```

```python
print(metrics.classification_report(y_test,y_pred))
```
